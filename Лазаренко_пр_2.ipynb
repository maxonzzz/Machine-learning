```python
import pandas as pd
import numpy as np

# Зчитайте дані з таблиці "Середньорічна щомісячна зарплата в Україні за 1992—2022 роки" з Вікіпедії за посиланням  та виконайте наступні завдання: 

url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

# 1. Виведіть перші 5 рядків таблиці. 

data = pd.read_html(url, header=0)[2]
df_1 = data.head(5)
print(df_1)
```

```python
# 2. Визначте кількість рядків і стовпчиків за допомогою атрибуту shape 

num_of_rows, numb_of_columns = data.shape
print(f"Кількість рядків: {num_of_rows}, Кількість стовпчиків: {numb_of_columns}")
```

```python
# 3. Визначте загальну кількість років та кількість місяців у даних. 
numb_of_years = len(data["Рік"])
print(numb_of_years)
numb_of_months = numb_of_years * 12
print(numb_of_months)
```

```python
# 4. Замінити пропущені значення  «-» в таблиці на NaN
data = data.replace("—", np.nan)
data
```

```python
# 5. Визначити тип стовпців за допомогою dtypes. 
column_types = data.dtypes
print("Типи стовпців:")
print(column_types)
```

```python
# 6. Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату,  Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею.  
data = data.drop(columns=["Кількість кгхліба[20] на середнюзарплату", "Кількість БігМаківна середню зарплату"])
data
```

```python
# 7. Замінити тип нечислових даних на числові.
#data["В національнійвалюті"] = data["В національнійвалюті"].astype(str)
#data["В доларах США"] = data["В доларах США"].astype(str)
data["В національнійвалюті"] = data["В національнійвалюті"].str.replace('▲', "").str.replace('▼', "")
data["В доларах США"] = data["В доларах США"].str.replace('▲', "").str.replace('▼', "")
data['В національнійвалюті'] = data['В національнійвалюті'].str.replace(',', '.').astype(float)
data['В доларах США'] = data['В доларах США'].str.replace(',', '.').astype(float)
print(data.dtypes)
```

```python
# 8. Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len) 
misses = data.iloc[:,1:].isnull().sum() / len(data)
print(misses)
```

```python
# 9. Замініть всі пропущені значення на середні значення по (стовпцю). 
for i, col in enumerate(data):
    data.iloc[:,i] = data.iloc[:,i].fillna(data.mean()) 
data
```

```python
# 10. Знайдіть рік з найвищою зарплатою в Україні в грн і вкажіть цю зарплату. 
hs_year = data['В національнійвалюті'].idxmax()
print("Pік з найвищою зарплатою: ", data["Рік"].iloc[hs_year])
```

```python
# 11. Знайдіть рік з мінімальною  зарплатою в Україні в дол. США і вкажіть цю зарплату. 
low_salary = data['В доларах США'].idxmin()
print("В доларах: ", data['В доларах США'].values[low_salary],"РіК: ", data["Рік"].iloc[low_salary])
```

```python
# 12. Виведіть роки, коли зарплата в грн була вище середньої
avg_salary = data['В національнійвалюті'].mean()
print(avg_salary)
above_avg = data[data['В національнійвалюті'] > avg_salary]
print(above_avg["Рік"].tolist())
```

```python
# 13. Порівняйте  зарплату  в грн за останні два доступних роки.  ДОРОБИТИ!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!

last_years = data.iloc[len(data)-2:len(data)]
print(last_years)
last_years_salary = last_years[last_years["В національнійвалюті"] == last_years["В національнійвалюті"]. min ()]
print("Нижча зарплата була в ", last_years_salary.iat[0,0], "році")
```

```python
# 14. Створіть графік, що відображає  зарплату в Україні з часом (за роками). 
import matplotlib.pyplot as plt
plt.figure(figsize=(25, 8))
plt.rcParams.update({'font.size': 14})
years = data['Рік']
salaries = data['В національнійвалюті']
plt.plot(years, salaries, marker='o', linestyle='-', color='b', label='Зарплата в гривнях')
plt.xlabel('Рік')
plt.ylabel('Зарплата в гривнях')
plt.title('Зарплата в Україні за роками')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 15. Видаліть останній рік з таблиці та виведіть оновлену таблицю.  
data = data.drop(data.index[-1])
data
```

```python
# 16. Побудуйте графік, що відображає середню зарплату в Україні в розрізі років без останнього року. 
average_salary = data.groupby('Рік')['В національнійвалюті'].mean()
plt.figure(figsize=(25, 6))
plt.plot(average_salary.index, average_salary.values, marker='o', linestyle='-', color='b')
plt.xlabel('Рік')
plt.ylabel('Середня зарплата в гривнях')
plt.title('Середня зарплата в Україні в розрізі років')
plt.grid(True)
plt.show()
```

```python
# 17. Обчисліть зарплату за останні два роки і порівняйте її із зарплатою за останні 5 років. Вкажіть, чи є різниця, і в якому напрямку. 

last_two_years = data.iloc[len(data)-2:len(data)]
last_five_years = data.iloc[len(data)-5:len(data)]
avg_twoy_salary = last_two_years["В національнійвалюті"].mean()
avg_fivey_salary = last_five_years["В національнійвалюті"].mean()
if avg_twoy_salary > avg_fivey_salary:
    print("Більша зарплатня була за останні два роки; Зарплатня: ", avg_twoy_salary, "Різниця: ", avg_twoy_salary - avg_fivey_salary )
elif avg_fivey_salary > avg_twoy_salary:
    print("Більша зарплатня була за останні п'ять років; Зарплатня: ", avg_fivey_salary, "Різниця: ", avg_fivey_salary - avg_twoy_salary )
else:
    print("Error")
```
