```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# 1. Прочитайте CSV файл (використовуйте метод read_csv):
data = pd.read_csv('Global YouTube Statistics.csv', encoding = 'unicode_escape')
```

```python
# 2. Виведіть перші п'ять рядків (використовується метод head):
df = data.head(5)
df
```

```python
# 3. Виведіть розміри датасету (використовуйте метод shape):
num_of_rows, numb_of_columns = data.shape
print(f"Кількість рядків: {num_of_rows}, Кількість стовпчиків: {numb_of_columns}")
```

```python
# 4. Перевірте, чи у всіх рядків вистачає даних: виведіть кількість пропусків  у кожному зі стовпців (використовуйте методи isna та sum).
#Якщо є пропуски, замінить їх на NaN, потім тип даних на числовий.
missing_values = data.isna().sum()
data.fillna(value="NaN")
print(missing_values)
new_data = data
for i in range(0, numb_of_columns):
    try:
        new_data.iloc[:i]-data.iloc[:i].astype(float).fillna(data.iloc[:i].mean)
    except:
        continue
new_data
```

```python
# 5. Замініть комірки з пропущеними значенями на середні за стовпцем. І тип даних на float.
new_data.fillna(new_data.mean(), inplace=True)
print(new_data.dtypes)
new_data
```

```python
# 6. Ще раз перевірте кількість пропусків.
missing_values = new_data.isna().sum()
print(missing_values)
```

```python
# 7. Перевірте, скільки унікальних країн представлено у колонці "Country"
unique_countries = new_data['Country'].unique()
count_unique_countries = len(unique_countries)
print(count_unique_countries)
```

```python
# 8. Побудуйте діаграму розподілу переглядів (використовуйте kind='hist'):
data['video views'].plot(kind='hist', bins=50, edgecolor='k', alpha=0.7)
plt.title('Розподіл переглядів')
plt.show()
```

```python
# 9. Визначте максимальну, мінімальну, середню кількість переглядів (використовуйте методи max, min, mean)
views = new_data['video views']
max_views = views.max()
min_views = views.min()
mean_views = views.mean()
print(f"Максимальна кількість переглядів: {max_views}")
print(f"Мінімальна кількість переглядів: {min_views}")
print(f"Середня кількість переглядів: {mean_views}")
```

```python
# 10.Знайдіть країну, де найбільше відео було завантажено на YouTube
most_uploads_country = data['Country'].value_counts().idxmax()
print(f'Країна з найбільшою кількістю завантажених відео на YouTube: {most_uploads_country}')
#ДОРОБИТИ
```
